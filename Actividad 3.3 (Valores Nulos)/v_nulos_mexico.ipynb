```python
#importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Cragamos archivo
data = pd.read_csv('México.csv')
data.head(5)
```

```python
#Identificar valores nulos por columna
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Eliminar columnas innecesarias
#axis = 1 para columnas
#axis = 0 para filas
data2 = data.drop(['description', 'host_thumbnail_url', 'minimum_minimum_nights', 'maximum_minimum_nights', 'minimum_maximum_nights', 'maximum_maximum_nights', 'minimum_nights_avg_ntm', 'maximum_nights_avg_ntm', 'availability_30', 'availability_60', 'availability_90', 'calendar_last_scraped', 'number_of_reviews_ltm', 'number_of_reviews_l30d', 'first_review', 'last_review', 'review_scores_accuracy', 'review_scores_checkin', 'review_scores_value', 'license', 'calculated_host_listings_count', 'calculated_host_listings_count_entire_homes', 'calculated_host_listings_count_private_rooms', 'calculated_host_listings_count_shared_rooms', 'neighbourhood_group_cleansed'], axis=1)
data2
```

```python
#Identificar valores nulos por columna
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
dataN=data2.copy()
```

```python
#Filtro para la columna: neighborhood_overview
#Sustituimos losvalores nulos por un string en concreto el cual es "The neighborhood is very nice" ya que al ser lugares turisticos estos siempre
#Procuran vender una buena imagen y ese suele ser un resumen de la opinion de muchos turistas
dataN["neighborhood_overview"] =dataN["neighborhood_overview"].fillna("The neighborhood is very nice")
```

```python
#Filtro para la variable host_location
#Se sustituyeron los valores nulos por valores no nulos hacia adelante ya que la localizacion era de la ciudad de mexico
#o conocido comel distritio federal (antes) Por que al colocar uno u otro en los valores nulos no implica un gran cambio en las variables
dataN["host_location"] =dataN["host_location"].fillna(method="ffill")
```

```python
#Filtro para la variable host_about
#Se ocupan los valores vacios con el ultimo valor valido ya que al tratarse de una pequeña presentacion del host
#En la cual estos mismos buscan dar una buena imagen aunque utlizen diferentes palabras.
dataN["host_about"] =dataN["host_about"].fillna(method="bfill")
```

```python
#Filtro para la variable host_response_time
#Ssustituir valores nulos por valores no nulos hacia adelante debido a que el tiempo de respuesta de los host no
#Suele tardar mas de un dia ((como lo indican las respuestas de la Base de datos))al tratarse de zonas de gran
#  interes para los turistas y esto no implica un cambio muy significativo en cuanto a lo que nos proporcionan los datos
dataN["host_response_time"] =dataN["host_response_time"].fillna(method="ffill")
```

```python
#Filtro para la variable host_response_rate
#Sustituimos los valores nulos con el promedio ya que al tratarse de una medida de tipo entero esto facilitaria
#el manejo de la variable y no se obtendrian conclusiones del todo erroneas

data2["host_response_rate"] = data2["host_response_rate"].str.rstrip('%').astype(float) / 100
dataN["host_response_rate"]=dataN["host_response_rate"].fillna(round(data2["host_response_rate"].mean(),1))
```

```python
#Metodo: Primer metodo
#Primero convertimos los datos de string a float, para despues donde hay datos nulos 
#para despues colocar el promedio 
data2["host_acceptance_rate"] = data2["host_acceptance_rate"].str.rstrip('%').astype(float) / 100
dataN["host_acceptance_rate"]=dataN["host_acceptance_rate"].fillna(round(data2["host_acceptance_rate"].mean(),1))
```

```python
#Metodo: Quinto metodo 
#En la variable host_is_superhost hicimos uso de ffill para completa el valor faltante con el último valor válido anterior 
dataN['host_is_superhost']=dataN['host_is_superhost'].fillna(method=("ffill"))
```

```python
#Metodo: Sexto metodo
#Debido a que en nuestro dataframe en  la columna de la variable "host_neighbourhood" hay varios datos nulos y no es posible  colocar información
#de celdas anteriores o posteriores ya que se estaría modificando información importante, por lo que se decidió sustituir 
#esas valores nulos por "Desconocido"
dataN['host_neighbourhood'] = dataN['host_neighbourhood'].fillna("Desconocido")
```

```python
#Metodo: Quinto metodo 
#En la variable neighborurhood hicimos uso de ffill para completa el valor faltante con el último valor válido anterior
dataN['neighbourhood']=dataN['neighbourhood'].fillna(method=("ffill"))
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en concreto
dataN["bathrooms"]=data2["bathrooms"].fillna(1)
dataN

#Descripcion: Se uso el tercer metodo de sustitucion para los registros de la columna "bathrooms", 
# pues este método nos ayuda a poner un numero en concreto, y colocamos que por lo menos todos tengan 1 baño 
```

```python
#Cuarto Método de sustitución de valores nulos
#Sustituir valores nulos por un string en concreto
dataN["bathrooms_text"] = dataN["bathrooms_text"].fillna("1 bath")

#Descripcion: se usa el cuarto método de sustitución de valores, en la columna "bathroom_text"
#pues esta columna contiene strings y se espera que por lo menos se cuente con un baño
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en concreto
dataN["bedrooms"]=data2["bedrooms"].fillna(1)
dataN

#Descripcion: Se uso el tercer método de sustitución para los registros de la columna "bedrooms", 
# pues este método nos ayuda a poner un número en concreto, y colocamos que por lo menos todos tengan 1 cuarto
```

```python
#Tercer método de sustitución de valores nulos
#Sustituir valores nulos por un valor numérico en concreto
dataN["beds"]=data2["beds"].fillna(1)
dataN

#Descripcion: Se uso el tercer método de sustitución para los registros de la columna "beds", 
# pues este método nos ayuda a poner un número en concreto, y colocamos que por lo menos todos tengan 1 cama
```

```python
#Cuarto Método de sustitución de valores nulos
#Sustituir valores nulos por un string en concreto
dataN["price"] = dataN["price"].fillna("$100")

#Descripcion: se usa el cuarto método de sustitución de valores, en la columna "price"
#pues esta columna contiene strings y se espera que por lo menos se cobre alguna comisión para la renta de un cuarto
```

```python
#Cuarto Método de sustitución de valores nulos
#Sustituir valores nulos por un string en concreto
dataN["calendar_updated"] = dataN["calendar_updated"].fillna("within data")

#Descripcion: se usa el cuarto método de sustitución de valores, en la columna "calendar_updated"
#pues esta columna contiene strings, y se informa que no contiene actualizaciones
```

```python
#Cuarto Método de sustitución de valores nulos
#Sustituir valores nulos por un string en concreto
dataN["has_availability"] = dataN["has_availability"].fillna("f")

#Descripción: se usa el cuarto método de sustitución de valores, en la columna "has_availability"
#pues esta columna contiene strings, y se informa que es falso o que no está habilitado o disponible
```

```python
#Segundo Método de sustitución de valores nulos
#Sustituir valores nulos con mediana
dataN["review_scores_rating"] = dataN["review_scores_rating"].fillna(round(data2["review_scores_rating"].median(),1)) #1 cuántos decimales quiero

#Descripción: se usa el segundo método de sustitución de valores nulos, pues esta sustitución nos ayuda con cuando hay valores atípicos o muy 
# variados, por ejemplo: puede haber reseñas de 1 a 5.
```

```python
#Segundo Método de sustitución de valores nulos
#Sustituir valores nulos con mediana
dataN["review_scores_cleanliness"] = dataN["review_scores_cleanliness"].fillna(round(data2["review_scores_cleanliness"].median(),1)) #1 cuántos decimales quiero

#Descripción: se usa el segundo método de sustitución de valores nulos, pues esta sustitución nos ayuda con cuando hay valores atípicos o muy 
# variados, por ejemplo: puede haber reseñas de 1 a 5.
```

```python
#Segundo Método de sustitución de valores nulos
#Sustituir valores nulos con mediana
dataN["review_scores_communication"] = dataN["review_scores_communication"].fillna(round(data2["review_scores_communication"].median(),1)) #1 cuántos decimales quiero

#Descripción: se usa el segundo método de sustitución de valores nulos, pues esta sustitución nos ayuda con cuando hay valores atípicos o muy 
# variados, por ejemplo: puede haber reseñas de 1 a 5.
```

```python
#Segundo Método de sustitución de valores nulos
#Sustituir valores nulos con mediana
dataN["review_scores_location"] = dataN["review_scores_location"].fillna(round(data2["review_scores_location"].median(),1)) #1 cuántos decimales quiero

#Descripción: se usa el segundo método de sustitución de valores nulos, pues esta sustitución nos ayuda con cuando hay valores atípicos o muy 
# variados, por ejemplo: puede haber reseñas de 1 a 5.
```

```python
#Uso del primer metodo de sustitucion de valores nulos,
#con este metodo se busca llenar los campos dionde hay datos nulos coen el promedio
#que en este caso seria el promedio de la variable reviews_per_mont
dataN['reviews_per_month']=dataN['reviews_per_month'].fillna(round(data2['reviews_per_month'].mean(),1))
```
